```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = [
    ["Random Forest", 96], 
    ["SVM", 91], 
    ["LSTM", 98], 
    ["Logistic Regression", 89] 
]
pd.DataFrame(df)
```

```python
df = pd.DataFrame(df, columns=["Model", "Accuracy"])

# Plot bar graph
plt.figure(figsize=(10, 6))
plt.bar(df["Model"], df["Accuracy"])
plt.title("Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy (%)")
plt.ylim(0, 100)
plt.xticks(rotation=20)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Your data
data = [
    ["Random Forest", 96], 
    ["SVM", 91], 
    ["LSTM", 98], 
    ["Logistic Regression", 89] 
]

df = pd.DataFrame(data, columns=["Model", "Accuracy"])

# Seaborn theme
sns.set_theme(style="whitegrid")

plt.figure(figsize=(4,6))

# Plot
ax = sns.barplot(data=df, x="Model", y="Accuracy")

# Add labels on top of bars
for i, v in enumerate(df["Accuracy"]):
    ax.text(i, v + 1, str(v), ha='center', fontweight='bold')

# Titles & labels
plt.title("Model Accuracy Comparison", fontsize=14)
plt.xlabel("Model")
plt.ylabel("Accuracy (%)")
plt.ylim(0, 105)
plt.xticks(rotation=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Data
data = {
    "Risk Score": [10, 20, 30, 40, 50, 60, 70, 80, 90],
    "Fraud Probability (%)": [2, 6, 12, 18, 26, 34, 43, 55, 70]
}

df = pd.DataFrame(data)

# Seaborn theme
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))

# Regression plot
sns.regplot(
    x="Risk Score",
    y="Fraud Probability (%)",
    data=df,
    scatter_kws={"s": 70},    # size of points
    line_kws={"linewidth": 2} # thickness of regression line
)

plt.title("Relationship Between Risk Score and Fraud Probability", fontsize=14)
plt.xlabel("Risk Score")
plt.ylabel("Fraud Probability (%)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = {
    "Month": [
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ],
    "Total Transactions (M)": [118.4, 120.1, 123.9, 127.8, 130.4, 132.7, 
                              135.2, 138.9, 140.5, 144.2, 147.8, 151.3],
    "Fraud Detected (Cases)": [2950, 3020, 3180, 3250, 3310, 3410,
                               3520, 3640, 3690, 3810, 3920, 4040]
}

df = pd.DataFrame(data)

# Style
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12,5))

# First axis
ax1 = plt.gca()
ax1.plot(df["Month"], df["Total Transactions (M)"], marker="o", linewidth=2, label="Total Transactions (M)")
ax1.set_xlabel("Month (2024)")
ax1.set_ylabel("Total Transactions (Millions)")
ax1.tick_params(axis='x', rotation=45)

# Second axis (twin)
ax2 = ax1.twinx()
ax2.plot(df["Month"], df["Fraud Detected (Cases)"], marker="s", linewidth=2, color="red", label="Fraud Detected (Cases)")
ax2.set_ylabel("Fraud Cases Detected")

# Title
plt.title("Monthly Transactions vs Fraud Cases Detected (2024)", fontsize=14)

# Combined legend
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
plt.legend(lines_1 + lines_2, labels_1 + labels_2, loc="upper left")

plt.tight_layout()
plt.show()
```
