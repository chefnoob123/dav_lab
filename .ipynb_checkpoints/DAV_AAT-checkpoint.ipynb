```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
df = pd.read_csv("Final-Data.csv")
df.head()
```

```python
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'], errors='coerce')
daily = df.groupby(df['Admission_Date'].dt.date).size().reset_index(name='Admissions')
daily['Admission_Date'] = pd.to_datetime(daily['Admission_Date'])
```

```python
plt.figure(figsize=(10,4))
plt.plot(daily['Admission_Date'], daily['Admissions'], marker='o')
plt.title("Daily Patient Admissions")
plt.xlabel("Admission Date")
plt.ylabel("Admissions")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter only occupied beds
occupied = df[df['Bed_Status'].str.lower() == 'occupied']

bed_counts = occupied.groupby('Department').size().reset_index(name='Occupied_Beds')
bed_counts = bed_counts.sort_values('Occupied_Beds', ascending=False)

plt.figure(figsize=(9,4))
plt.bar(bed_counts['Department'], bed_counts['Occupied_Beds'])
plt.title("Bed Occupancy by Department")
plt.xlabel("Department")
plt.ylabel("Occupied Beds")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
outcomes = df['Treatment_Outcome'].value_counts()

fig, ax = plt.subplots(figsize=(6,6))
ax.pie(outcomes.values, labels=outcomes.index, autopct='%1.1f%%', startangle=90)

# donut hole
circle = plt.Circle((0,0), 0.40, color='white')
fig.gca().add_artist(circle)

plt.title("Treatment Outcomes")
plt.tight_layout()
plt.show()
```

```python
pivot = df.groupby([df["Admission_Date"].dt.date, "Department"]).size().unstack(fill_value=0)

plt.figure(figsize=(12,5))
sns.heatmap(
    pivot.T,
    annot=True,
    fmt="d",
    cmap="Blues",
    linewidths=0.5,      
    linecolor="black"    
)

plt.title("Heatmap: Patient Count by Department and Admission Date")
plt.xlabel("Admission Date")
plt.ylabel("Department")
plt.tight_layout()
plt.show()
```
